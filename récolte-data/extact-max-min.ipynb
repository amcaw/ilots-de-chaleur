```python
import pandas as pd 
import geopandas as gpd
```

```python
final_df = pd.DataFrame()

cities = ["charleroi", "liege", "mons", "namur", "tournai"]
for city in cities : 

    df = gpd.read_file(f"../assets/selection/{city}_sel_centroid.json")
    df.dropna(inplace=True)
    df.reset_index(inplace=True)
    maxIn = df["raster_value"].idxmax()
    minIn = df["raster_value"].idxmin()
    df['ville'] = city
    df_max = df.iloc[[maxIn]]
    df_max["MINMAX"] = "Max"
    df_min = df.iloc[[minIn]]
    df_min["MINMAX"] = "Min"
    final_df = pd.concat([final_df,df_max], ignore_index=True)
    final_df = pd.concat([final_df, df_min], ignore_index=True)

display(final_df)
final_df.to_file("../data/min_max_every_city.json", driver="GeoJSON")


```
