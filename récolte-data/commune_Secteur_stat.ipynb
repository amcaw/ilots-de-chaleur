```python
import pandas as pd 
import geopandas as gpd
```

```python
stat = gpd.read_file("sh_statbel_statistical_sectors_31370_20230101.json")
stat.head()
```

```python
Liege = "62063"
charleroi = "52011"
mons = "53053"
namur = "92094"
tournai = "57081"
```

```python
liege_gpd = stat.loc[stat["cd_munty_refnis"] == Liege]
charleroi_gpd = stat.loc[stat["cd_munty_refnis"] == charleroi]
mons_gpd = stat.loc[stat["cd_munty_refnis"] == mons]
namur_gpd = stat.loc[stat["cd_munty_refnis"] == namur]
tournai_gpd = stat.loc[stat["cd_munty_refnis"] == tournai]
```

```python
liege_gpd.to_file("./secteurs/liege_secteur.geojson", driver="GeoJSON")
charleroi_gpd.to_file("./secteurs/charleroi_secteur.geojson", driver="GeoJSON")
mons_gpd.to_file("./secteurs/mons_secteur.geojson", driver="GeoJSON")
namur_gpd.to_file("./secteurs/namur_secteur.geojson", driver="GeoJSON")
tournai_gpd.to_file("./secteurs/tournai_secteur.geojson", driver="GeoJSON")
```
