```python
import geopandas as gpd
import pandas as pd
```

```python
cities = ["namur", "liege", "tournai", "charleroi", "mons"]
every = pd.DataFrame()
for city in cities:
    test = gpd.read_file(f"../data/secteurs/{city}_secteur.geojson")

    test["city"] = city
    every = pd.concat([every, test], ignore_index=True)

every.to_file(f"../data/secteurs/every_city.json", driver="GeoJSON")
every
```
